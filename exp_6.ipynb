```python
#6i. Salary Calculation with Allowances

'''Problem Statement: Write a Python program to calculate the gross salary of an employee

if basic <= 10000: HRA = 20%, DA = 80%
if basic <= 20000: HRA = 25%, DA = 90%
otherwise: HRA = 30%, DA = 95%'''

salary = int(input("Enter your Basic Salary:"))

if salary <= 10000:
  gross = 0.2*salary + 0.8*salary + salary
elif salary <= 20000:
  gross = 0.25*salary + 0.9*salary + salary
else:
  gross = 0.3*salary + 0.95*salary + salary

print("Your Gross Salary is: ", gross)
```

```python
#6j. Write a Python program to Calculate Income Tax based on Annual Income:

'''Up to Rupees 2,50,000 : No Tax
Rupees 2,50,001 - 5,00,000 : 5%
Rupees 5,00,001 - 10,00,000 : 20%
Above Rupees 10,00,000 : 30%'''

income=float(input("Enter Gross Annual Income: "))

if income <= 250000:
  tax = 0
elif income <= 500000:
  tax = (income-250000)*0.05
elif income <= 1000000:
  tax = (250000*0.05) + (income - 500000)*0.2
else:
  tax = (250000*0.05) + (500000*0.2) + (income - 1000000)*30

print("Your Payable Income Tax is: ", tax)
```
